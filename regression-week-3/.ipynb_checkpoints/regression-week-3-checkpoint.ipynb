## Polynomial Regression

```python
import math
import numpy as np
import pandas as pd
from sklearn import linear_model
```

## Importing Sales data

```python
sales_data = pd.read_csv('../kc_house_data.csv')
sales_data = sales_data.sort_values(['sqft_living', 'price'])
train_data = pd.read_csv('wk3_kc_house_train_data.csv')
valid_data = pd.read_csv('wk3_kc_house_valid_data.csv')
test_data = pd.read_csv('wk3_kc_house_test_data.csv')
set_1_data = pd.read_csv('wk3_kc_house_set_1_data.csv')
set_2_data = pd.read_csv('wk3_kc_house_set_2_data.csv')
set_3_data = pd.read_csv('wk3_kc_house_set_3_data.csv')
set_4_data = pd.read_csv('wk3_kc_house_set_4_data.csv')
```

```python
def polynomial_sframe(feature, degree):
    poly_sframe = pd.DataFrame()
    poly_sframe['power_1'] = feature;
    if degree > 1:
        for power in range(2, degree + 1):
            name = 'power_'+ str(power)
            poly_sframe[name] = feature.apply(lambda x: x ** power)
    return poly_sframe
```

## First Degree Polynomial

```python
poly1_data = polynomial_sframe(sales_data['sqft_living'], 1)
poly1_data['price'] = sales_data['price']

model1 = linear_model.LinearRegression()
model1.fit(poly1_data.drop('price', axis=1), poly1_data['price'])

import matplotlib.pyplot as plt
%matplotlib inline
plt.plot(poly1_data['power_1'], poly1_data['price'], '.',
        poly1_data['power_1'], model1.predict(poly1_data.drop('price', axis=1)), '-')
```

```python
rss = []
for degree in range(1, 16):
    poly_data = polynomial_sframe(train_data['sqft_living'], degree)
    poly_data['price'] = train_data['price']

    # fit model on the current degree     
    model = linear_model.LinearRegression()
    model.fit(poly_data.drop('price', axis=1), poly_data['price'])
    
    # compute the polynomial sframe for the validation data
    poly_validation_data = polynomial_sframe(valid_data['sqft_living'], degree)
    poly_validation_data['price'] = valid_data['price']
    
    # predict the prices on the training data using the current model
    # and compute the rss.
    rss_value = ((poly_validation_data['price'] - model.predict(poly_validation_data.drop('price', axis=1))) ** 2).sum()
    rss.append((degree, rss_value))
sorted(rss, key=(lambda x: x[1]))
```

## Compute RSS on the Test Data

```python
poly_data = polynomial_sframe(train_data['sqft_living'], 6)
poly_data['price'] = train_data['price']

# fit model on the current degree     
model = linear_model.LinearRegression()
model.fit(poly_data.drop('price', axis=1), poly_data['price'])

# compute the polynomial sframe for the validation data
poly_test_data = polynomial_sframe(test_data['sqft_living'], 6)
poly_test_data['price'] = test_data['price']

# predict the prices on the training data using the current model
# and compute the rss.
rss_value = ((poly_test_data['price'] - model.predict(poly_test_data.drop('price', axis=1))) ** 2).sum()
print(rss_value)
```
